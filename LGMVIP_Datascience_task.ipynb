```
from google.colab.patches import cv2_imshow
```

```
import cv2
```

```
img=cv2.imread('cat.jpg')
res=cv2.resize(img,(300,300))
cv2_imshow(res)
cv2.waitKey(0)
```

```
gray=cv2.cvtColor(res,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
cv2.waitKey(0)
```

```
inverted=255-gray
cv2_imshow(inverted)
cv2.waitKey(0)
```

```
blurred=cv2.GaussianBlur(inverted,(21,21),0)
cv2_imshow(blurred)
cv2.waitKey(0)
```

```
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray,inverted_blurred,scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

```
cv2_imshow(res)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
